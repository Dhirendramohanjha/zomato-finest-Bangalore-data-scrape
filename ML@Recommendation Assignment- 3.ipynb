```python
train_data, test_data = train_test_split(data, test_size = 0.20, random_state=100)
train_data.head(5)
```

```python
from Recommender import item_similarity_recommender_py
is_model =item_similarity_recommender_py()
is_model.create(train_data, 'UserId', 'Restro')
```

```python
#Print the Restraurants for the user in training data
UserId= users[4]
user_items = is_model.get_user_items(UserId)
#
print("------------------------------------------------------------------------------------")
print("Training data songs for the user userid: %s:" % UserId)
print("------------------------------------------------------------------------------------")

for user_item in user_items:
    print(user_items)

print("----------------------------------------------------------------------")
print("Recommendation process going on:")
print("----------------------------------------------------------------------")

#Recommend Restraurants for the user using personalized model
is_model.recommend(UserId)
```
