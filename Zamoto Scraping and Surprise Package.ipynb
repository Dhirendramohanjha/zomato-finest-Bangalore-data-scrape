## 2. Recommendation engine using User Based Collaborative Filtering

```python
import surprise
from surprise import KNNWithZScore
from surprise import Reader, Dataset
import pandas as pd
import numpy as np
```

```python
zomato = pd.read_csv('zomato_user_ratings.csv')
zomato.shape, zomato['UserId'].nunique()
```

### As 1 user can give multiple ratings to same restro, we are taking average of ratings for 1 user for 1 restro

```python
z_df = zomato.groupby(['UserId','Restro_name']).agg({'Rating':np.mean})
z_df = z_df.reset_index()
z_df.shape, z_df['UserId'].nunique()
```

```python
z_df['Rating'].max()
```

```python
reader = Reader(rating_scale = (1,5))

data = Dataset.load_from_df(z_df,reader)

sim_options = {
    'name': 'cosine',
    'user_based': True,
    'k': 5    
}
knn = KNNWithZScore(sim_options = sim_options)
```

```python
knn.fit(data.build_full_trainset())
```

```python
z_df[z_df.UserId==87390]
```

### Predicting for UserId = 87390

```python
userid = 87390
predictions = pd.DataFrame()

for restro_name in z_df['Restro_name'].unique():
    restro_row = z_df[(z_df['UserId'] == userid) &
                         (z_df['Restro_name'] == restro_name)]
    
    if len(restro_row) == 0:
        restro_pred_rating = knn.predict(uid=userid, iid = restro_name)[3]
        predictions = predictions.append({
            'Restro_name':restro_name,
            'rating_predictions': restro_pred_rating
        }, ignore_index = True)
predictions.head()
```
